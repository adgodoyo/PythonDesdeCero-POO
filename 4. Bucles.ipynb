# FOR bucle finito: "Por cada cosa, haré..."

Los humanos siempre buscamos algo que se llama "economía de la lenguage", un proceso evolutivo enfocado a la minimización del esfuerzo invertido a la hora de hablar, escribir o expresarse en general. Un ejemplo de ello, es el uso de *bucles*.

Imagina que tienes 20 invitados a una fiesta que organizaste y quieres darles un vaso de gaseosa, y para ello le pedirás el favor a un amigo que te ayude a servirlos. Darle la siguiente instrucción sería muy ineficiente: 

"toma un vaso, sirve gaseosa y pásaselo al invitado 1, luego toma un vaso, sirve gaseosa y pásaselo al invitado 2, luego, toma un vaso, sirve gaseosa y pásaselo al invitado 3,....toma un vaso, sirve gaseosa y pásaselo al invitado 20"

Una forma más simple es con palabras que transmiten la idea de bucles:

"**Por cada** __invitado__ en __la fiesta__, toma un vaso, sirve gaseosa y pásalo"



Por cada, señala que vamos a repetir acciones, ¿cuántas veces? tantos invitados haya en la fiesta, es decir, habrá x repeticiones como x **objetos** haya en un **grupo**.

Nuestro cerebro ya está entrenado para entender que para saber el número de repeticiones necesito saber qué **elemento** dentro de qué **conjunto** (siempre esas dos)



En Python funciona igual:

```python
invitados=["Juan", "Maria", "Carlos", "Pedro", "Hernesto", "Fernanda", "Daniel", "Andrés", "Luisa", "Martha", "Lennon", "McCartney", "Ringo", "Harrison"]

#para cada elemento en el grupo:
for persona in invitados:
    print("Toma un vaso")
    print("Sirve gaseosa")
    print(f"Entrega a {persona}")
    print("Vuelve a la cocina")
```

```python
persona
```

### Sintaxis

1. Empieza con **for**, que es como decir "Para cada".
2. Luego especificas una **variable(una caja)** que va a cambiar su **valor(objeto)** con cada iteración del bucle, irá tomando sucesivamente cada objeto del grupo
3. Después usas la palabra **in**, que es como decir "en".
4. Luego nombras el **grupo** (lista, tupla, diccionario, fila, columna, base de datos, etc.) sobre la que estás iterando
5. terminas la línea con **":"**, que abre paso a un bloque de código, donde explicas la acción que realizarás, equivalente a "voy a realizar esta acción".
6. Luego con sangría (tab), colocas los comandos a repetir

```python
# Ejemplos
# 1. Procesar una lista de empresas:

empresas = ["Empresa A", "Empresa B", "Empresa C"]
for item in empresas: 
    print(item)
```

```python
# 4. Lista de empresas con su participación de mercado
empresas = [
    {'nombre': 'Empresa A', 'participacion': 40},
    {'nombre': 'Empresa B', 'participacion': 20},
    {'nombre': 'Empresa C', 'participacion': 15},
    {'nombre': 'Empresa D', 'participacion': 15},
    {'nombre': 'Empresa E', 'participacion': 10}
]

# Usamos un bucle for para iterar sobre la lista de empresas
for empresa in empresas:
    participacion = empresa['participacion']
    # Verificamos si alguna empresa tiene más del 40% del mercado, lo cual puede indicar una posición dominante
    if participacion >= 40:
        print(f"Alerta: {empresa['nombre']} tiene una posición dominante con una participación de mercado del {participacion}%.")
    else:
        print(f"{empresa['nombre']} tiene una participación de mercado del {participacion}%. Dentro de los parámetros normales.")
```

# While bucle indeterminado: "Mientras una condición sea cierta, algo se repite"

While al igual que For es un bucle que señala que una(s) acción(es) debe ser repetida. En el caso de for, las repeticiones están determinadas por el número de objetos en un conjunto. While, por otro lado, no considera ni objetos ni elementos, considera solo una condición: **Mientras** algo sea cierto (True - boolean), una acción se repetirá, cuando la condición se vuelva False, no se repite más.

"Piensa en un while como una fiesta en la que el DJ está poniendo tu música favorita. La instrucción que le das a tu cuerpo es bastante simple: 'Mientras la música que suena sea de mi agrado, seguiré bailando'. No tienes un número definido de canciones que quieres bailar; simplemente, tu acción de bailar está condicionada a la presencia de esa música que te gusta.

En el mundo de la programación, el bucle while sigue una lógica similar. Le proporcionamos a nuestro programa una condición y le decimos: 'Mientras esta condición sea verdadera, sigue ejecutando las instrucciones que te he dado'.



```python
musica_suena=True
while musica_suena:
    print("Dar vuelta")
```

```python
buena_musica=True
while buena_musica:
    print("bailar")
```

# Funciones: rutinas que se ejecutan con solo decir una palabra

Una función te permite encapsular un bloque de código para que sea reutilizable y puedas ejecutarlo muchas veces sin tener que volverlo a escribir. En la vida real, cuando alguien te pide que alces la mano, no necesitas que te detallen cada pequeño movimiento muscular necesario para seguir la instrucción. Sabes que "alzar la mano" encapsula una serie de movimientos coordinados que has aprendido y no es necesario que los mencionen todos. Tampoco es necesario que te especifiquen los cientos de movimientos que se deben requerir para "preparar el desayuno". Todo el tiempo creamos funciones, aprendemos las intrucciones y el paso a paso una vez, y luego le ponemos a esa rutina un nombre.



En python funciona así:

```python
def alzar_la_mano():    
    print("Rotar el hombro para posicionar el brazo lateralmente.")
    print("Contraer los músculos del hombro para iniciar el levantamiento del brazo.")
    print("Continuar levantando el brazo hasta que esté horizontal al suelo.")
    print("Estirar completamente el codo para que el brazo quede recto.")
    print("Girar la muñeca para que la palma de la mano mire hacia adelante.")
    print("Separar los dedos para abrir la mano.")
    print("listo mano levantada")
    print("dí hola")
```

```python
alzar_la_mano()
```

```python
def calcular_HHI(cuotas_mercado): #Cuotas_mercado debe ser lista o tupla
    """
    Calcula el Índice de Herfindahl-Hirschman (HHI) para un conjunto dado de cuotas de mercado.

    Parámetros:
    cuotas_mercado (list): Una lista de cuotas de mercado de las firmas en el mercado. 
                           Cada cuota está representada como un porcentaje (ej. 25 para un 25%).

    Retorna:
    int: El valor del HHI calculado.
    """
    hhi = 0
    for cuota in cuotas_mercado:
        cuadrado = cuota ** 2
        hhi = hhi + cuadrado
    return hhi
```

```python
# Ejemplo de uso:
cuotas1 = [10, 20, 30, 40]  # Representa las cuotas de mercado de 4 firmas en porcentaje

print(f"El Índice de Herfindahl-Hirschman es: {calcular_HHI(cuotas1)}")
```

### Return 
Para explicar **return** en el contexto de una función en Python, podríamos extender la analogía de una máquina. Imagina que la función es como una máquina en una fábrica o una persona realizando un encargo, y return es el resultado final o el producto que te entregan al final del proceso.

Por ejemplo, cuando pides un café en una cafetería, hay una serie de pasos que el barista sigue para hacerlo: moler los granos de café, poner el café en la máquina, presionar el agua a través del café, etc. No necesitas saber todos los detalles cuando haces tu pedido; solo esperas obtener una taza de café al final.

En términos de programación, la función sería el proceso de hacer el café, y return sería el OBJETO, puede ser un objeto string, entero, float, boolean o de otros tipos. Al final, la función será un objeto.
